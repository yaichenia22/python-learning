## Why Type Hints?As a Java developer, you're used to the compiler catching type errors. Python's `mypy` gives you the same safety, without the verbosity.

````python
from typing import Optional, Unionfrom collections.abc import Sequence# Basic type hintsdef process_record(    record_id: int,    name: str,    score: float,    tags: list[str] | None = None,   # Python 3.10+ union syntax) -> dict[str, any]:    return {        "id": record_id,        "name": name.upper(),        "score": round(score, 2),        "tags": tags or [],    }result = process_record(1, "patient_a", 98.567, ["critical", "icu"])print(result)
````

````python
from dataclasses import dataclass, fieldfrom datetime import datetime# Dataclasses — like Java records / Lombok @Data@dataclassclass PipelineRun:    pipeline_name: str    status: str  # "running" | "success" | "failed"    started_at: datetime = field(default_factory=datetime.now)    records_processed: int = 0    errors: list[str] = field(default_factory=list)        @property    def is_successful(self) -> bool:        return self.status == "success" and not self.errors# Auto-generates __init__, __repr__, __eq__run = PipelineRun("etl_healthcare", "success", records_processed=50_000)print(run)print(f"Successful: {run.is_successful}")
````

## Protocols — Structural Typing☕ **Java parallel:** Like interfaces, but without `implements`. Duck typing with safety.

````python
from typing import Protocol, runtime_checkable@runtime_checkableclass DataSource(Protocol):    """Any object with a read() method is a DataSource."""    def read(self) -> list[dict]: ...    def close(self) -> None: ...class DatabaseSource:    def read(self) -> list[dict]:        return [{"id": 1, "value": "data"}]    def close(self) -> None:        print("DB connection closed")class FileSource:    def read(self) -> list[dict]:        return [{"id": 2, "value": "file_data"}]    def close(self) -> None:        print("File handle closed")def ingest(source: DataSource) -> list[dict]:    """Works with ANY object that has read() and close()."""    try:        return source.read()    finally:        source.close()# No 'implements DataSource' needed — just matching methodsprint(isinstance(DatabaseSource(), DataSource))  # True at runtime!print(ingest(DatabaseSource()))
````
