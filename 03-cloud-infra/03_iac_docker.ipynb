## Dockerfile for Python Data Apps

```python
# %%writefile Dockerfile.example# Multi-stage build for a Python data pipelinedockerfile_content = '''# Stage 1: BuildFROM python:3.11-slim AS builderWORKDIR /appCOPY requirements.txt .RUN pip install --no-cache-dir --prefix=/install -r requirements.txt# Stage 2: RuntimeFROM python:3.11-slimWORKDIR /appCOPY --from=builder /install /usr/localCOPY . .# Non-root user (security best practice)RUN useradd --create-home appuserUSER appuserENTRYPOINT ["python", "-m", "pipeline.main"]'''print(dockerfile_content)
```
