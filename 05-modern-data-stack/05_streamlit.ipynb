```python
# Save as dashboard.py and run with: streamlit run dashboard.pystreamlit_code = '''import streamlit as stimport pandas as pdimport numpy as npst.set_page_config(page_title="Healthcare Analytics", layout="wide")st.title("Healthcare Pipeline Dashboard")# Sidebar filtersst.sidebar.header("Filters")diagnosis = st.sidebar.multiselect(    "Diagnosis",    ["diabetes", "copd", "heart_failure", "pneumonia"],    default=["diabetes", "copd"],)age_range = st.sidebar.slider("Age Range", 18, 90, (30, 70))# Generate sample datanp.random.seed(42)df = pd.DataFrame({    "patient_id": [f"P{i:04d}" for i in range(200)],    "age": np.random.randint(18, 90, 200),    "diagnosis": np.random.choice(["diabetes", "copd", "heart_failure", "pneumonia"], 200),    "cost": np.random.lognormal(8, 0.8, 200).round(2),    "readmitted": np.random.choice([True, False], 200, p=[0.3, 0.7]),})# Apply filtersfiltered = df[    (df["diagnosis"].isin(diagnosis)) &    (df["age"].between(*age_range))]# KPI cardscol1, col2, col3, col4 = st.columns(4)col1.metric("Total Patients", len(filtered))col2.metric("Avg Cost", f"${filtered[\'cost\'].mean():,.0f}")col3.metric("Readmission Rate", f"{filtered[\'readmitted\'].mean():.1%}")col4.metric("Avg Age", f"{filtered[\'age\'].mean():.0f}")# Chartsst.subheader("Cost Distribution by Diagnosis")st.bar_chart(filtered.groupby("diagnosis")["cost"].mean())st.subheader("Patient Data")st.dataframe(filtered, use_container_width=True)'''print(streamlit_code)print("\n# Run: streamlit run dashboard.py")
```
